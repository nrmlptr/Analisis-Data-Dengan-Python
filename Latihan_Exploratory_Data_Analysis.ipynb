Eksplorasi Data orders_df
mengandung berbagai informasi terkait sebuah order yang terdiri dari order_id, customer_id, order_date, dan delivery_date.
berdasarkan informasi tersebut, dapat dibuat kolom baru untuk memuat waktu pengiriman tiap order. untuk melakukannya, perlu menghitung selisih antar delivery_date dan order_date serta menyimpannya sebagai delivery_time.
lalu, gunakan method apply() untuk melakukan sebauh operasi terhadap setiap elemen dalam sebuah kolom dataframe atau series. Operasi yang akan dilakukan yaitu menghitung jumlah detik dari delivery_time dengan method total_seconds(). Nilai tersebut nantinya diubah kedalam satuan hari (dibagi 86400) dan diambil bilangan bulatnya saja.

```python
orders_df.sample(5)
```

```python
delivery_time = orders_df["delivery_date"] - orders_df["order_date"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
orders_df["delivery_time"] = round(delivery_time/86400)
```

```python
orders_df.describe(include="all", datetime_is_numeric=True)
```

Berdasarkan hasil tersebut, dapat diketahui bahwa rata-rata waktu pengiriman sebesar 14 hari dengan nilai maksimum sebesar 27 hari dan nilai minimum sebesar 1 hari.

Eksplorasi Data orders_df dan customers_df

pada data orders_df terdapat kolom yang berisi informasi terkait customer id pelanggan yang pernah melakukan order. Informasi ini bisa digunakan untuk mengidentifikasi pelanggan yang belum pernah melakukan order. Untuk melakukan hal ini, bisa membuat sebuah kolom baru bernama “status” pada data customers_df. Kolom tersebut memiliki nilai “Active” untuk pelanggan yang pernah melakukan order setidaknya sekali dan sebaliknya bernilai “Non Active” untuk pelanggan yang belum pernah melakukan order sama sekali.

```python
customer_id_in_orders_df = orders_df.customer_id.tolist()
customers_df["status"] = customers_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_orders_df else "Non Active")
customers_df.sample(5)
```

untuk memperoleh inforamsi jumlah pelanggan yang berstatus aktif dan tidak aktif bisa gunakan pivot table.

```python
customers_df.groupby(by="status").customer_id.count()
```

dari hasil diatas, terdapat cukup banyak pelanggan yang belum pernah melakukan transaksi sebelumnya. hampir 30% pelanggan belum pernah melakukan order sebelumnya. untuk memperoleh informasi lebih banyak terkait 2 data tsb, perlu gabungkan keduanya melalui proses merge atau join.

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)

orders_customers_df.head()
```

pada kode diatas, proses merge dengan metode "left" dilakukan untuk mengambil semua nilai dari tabel kiri serta nilai yang bersesuaian dengan tabel kanan.
terdapat banyak sekali informasi yang bisa di eksplore.

```python
# jumlah order berdasarkan kota
orders_customers_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

dari data diatas, diketahui kota jordanside dan new ava merupakan kota yang memiliki jumlah order terbanyak.

```python
# jumlah order berdasarkan state
orders_customers_df.groupby(by="state").order_id.nunique().sort_values(ascending=False)
```

diketahui dari data diatas bahwa south australia merupakan state yang paling banyak melakukan order.

```python
# jumlah order berdasarkan gender
orders_customers_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False)
```

berdasarkan hasil tersebut, diketahui kebanyakan order dilakukan oleh pelanggan yang berjenis kelamin prefer not to say. sejalan dengan jumlah pelanggan yang kebanyakan dari kelompok gender tsb.

```python
# jumlah order berdasarkan kelompok usia
# definisikan kolom baru bernama age_group
# kelompokan pelanggan jadi 3 kelompok (youth, adults, seniors)
orders_customers_df["age_group"] = orders_customers_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
orders_customers_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

berdasarkan data diatas, diketahui bahwa pelanggan paling banyak melakukan order adalah kelompok usia adults.

perhatikan, trousers adalah tipe produk yang paling laris terjual. tapi jika diliaht berdasarkan revenue yang diterima, jacket merupakan tipe produk yang paling banyak menyumbang revenue ke perusahaan.

```python
# buat pivot table untuk informasi penjualan based on nama produk

sales_product_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

dari hasil data diatas, diketahui bahwa produk denim merupakan produk yang paling laris dan paling banyak menyumbang revenue ke perusahaan.
