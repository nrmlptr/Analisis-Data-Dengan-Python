histogram

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(15, 5, 250)

plt.hist(x=x, bins=15)
plt.show()
```

```python
# membuat grafi histogram dengan seaborn
import seaborn as sns

sns.histplot(x=x, bins=15)
plt.show()
```

```python
# gunakan kde=True untuk permudah identifikasi distribusi dari data kuantitatif pada histogram
import seaborn as sns

sns.histplot(x=x, bins=15, kde=True)
plt.show()
```
