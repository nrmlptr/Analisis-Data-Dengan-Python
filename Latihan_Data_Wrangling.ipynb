 selanjutnya kita ada dalam proses terakhir dalam data wrangling yaitu pembersihan atau cleaning data. Pada tahap ini, kita akan membersihkan berbagai masalah yang telah teridentifikasi dalam proses assessing data.

Membersihkan Data customers_df
Berdasarkan hasil proses assessing data, diketahui bahwa terdapat tiga masalah yang dijumpai dalam customer_df, yaitu duplicate data, missing value, dan inaccurate value.

```python
# Menghilangkan duplicate data
customers_df.drop_duplicates(inplace=True)
```

```python
# periksa kembali duplikasi data
print("Jumlah duplikasi: ", customers_df.duplicated().sum())
```

```python
# Menangani missing value
# gunakan teknik filtering untuk mengetahui data yang mengandung missing value
customers_df[customers_df.gender.isna()]
```

karena banyak informasi pada data yang sayang jika dibuang, maka gunakan metode imputation. Pada metode imputation, akan menggunakan nilai tertentu untuk mengganti missing value. Musabab kolom gender merupakan kolom kategorik, gunakan nilai yang dominan sebagai pengganti missing value tersebut. Gunakanlah method value_counts()untuk mengidentifikasi nilai yang dominan.

```python
# identifikasi nilai dominan
customers_df.gender.value_counts()
```

```python
# gunakan nilai dominan untuk mengganti missing value
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
# periksa kembali missing value
customers_df.isna().sum()
```

```python
# Menangani inaccurate value
# melihat data baris data yang mengandung inaccurate value tersebut
customers_df[customers_df.age == customers_df.age.max()]
```

Berdasarkan data tersebut, asumsikan bahwa inaccurate value tersebut terjadi karena human error sehingga kelebihan memasukkan nilai nol. Oleh karena itu, gantilah dengan nilai 70. Proses ini dilakukan dengan memanfaatkan method replace()

```python
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
# periksa kembali baris data yang inacurrate
customers_df[customers_df.age == customers_df.age.max()]
```

```python
# perbaiki kembali data
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
```

```python
# periksa apakah masih ada inaccurate data?
customers_df.describe()
```

Berdasarkan hasil tersebut dapat dilihat bahwa kolom age memiliki nilai maksimum yang cukup masuk akal. Selain itu, perhatikan, nilai mean dan standard deviation ikut berubah setelah kasus inaccurate value ditangani.

Selanjutnya Membersihkan Data orders_df.
Berdasarkan proses assessing data sebelumnya, diketahui bahwa terdapat kesalahan tipe data untuk kolom order_date & delivery_date.

Untuk mengatasi masalah ini, ganti tipe data pada kolom order_date & delivery_date menjadi datetime. Proses ini dapat dilakukan menggunakan function to_datetime() yang disediakan oleh library pandas.

```python
# ubah tipe data menjadi datetime
datetime_columns = ["order_date", "delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
# periksa kembali tipe data dan jumlah data
orders_df.info()
```

Selanjutnya membersihkan Data sales_df
Berdasarkan hasil penilaian data sebelumnya, diketahui bahwa terdapat 19 missing value pada kolom total_price. Untuk mengetahui proses penanganan missing value yang paling sesuai,perlu melihat terlebih dahulu baris data yang mengandung missing value tersebut.

```python
sales_df[sales_df.total_price.isna()]
```

Berdasarkan tampilan data tersebut, ditemukan bahwa nilai total_price merupakan hasil perkalian antara price_per_unit dan quantity. maka dapat gunakan pola ini untuk menangani missing value pada kolom total_price.

```python
sales_df["total_price"] = sales_df["price_per_unit"] * sales_df["quantity"]
```
